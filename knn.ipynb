```python
import numpy as np;
```

```python
#step 2:
data = np.array([[25, 45],      #Low splender
                [35, 70],       #High splender
                [45, 50],
                [32, 85],
                [28, 65]])
```

```python
labels = np.array([0,1,0,1,1])
```

```python
# step 3: Define the new coustomer to classify (Age = 30, Income = 60)
new_coustomer = np.array([30, 60])
```

```python
# step 4: Function to calculate Euclidean Distance
def euclidean_distance(point1, point2):
    return np.sqrt(np.sum((point1 - point2)**2))
```

```python
# step 5: Mannually compute distances from new coustomers to each existing coustomers
distances =[]
for i in range(len(data)):
    dist = euclidean_distance(new_coustomer, data[i])
    distances.append((dist,labels[i]))                 # store distance and corresponding class
```

```python
# step 6: sort the distance by distance value
distances.sort(key= lambda x:x[0])
```

```python
# step 7: select k = 3 nearest neighbour
k =3
nearest_neighbors = distances[:k]
```

```python
print(nearest_neighbors)
```

```python
#step 8 :Perform Majority voting to determine the class
votes = [neighbor[1] for neighbor in nearest_neighbors]
predicted_class = max(set(votes), key = votes.count) # majority vote
```

```python
# output the result
if predicted_class == 0:
    print("The new coustomer is classified as : Low Splender")
else:
    print("The new coustomer is classified as : High Splender")
```

```python
# (Optional) print the distances and neighbors for calrity
print("\nDistances from the new coustomers to existing coustomers")

for i, (dist, labels) in enumerate(distances):
    print(f"Coustomer {i+1}: Distance = {dist:.2f}, Label = {labels}")
```
