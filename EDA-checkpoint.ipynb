```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
         'Age' : [12,23,34,22,24,24],
'Height' : [165,145,170,146,198,180],
    'Weight' : [40,45,67,45,56,59],
    'Gender': ['M','F','M','F','F','F'],
    'Smoker' : ['No','Yes','Yes','Yes','Yes','No']
}
```

```python
df =  pd.DataFrame(data)
```

```python
df
```

# Understand the data

```python
print(df.describe())
df.dtypes
```

```python
plt.figure(figsize=(10,4))
```

```python
plt.subplot(1,2,1)
sns.histplot(df['Age'],kde = True, color='blue')
plt.title('Age distribution (discreet data)')
```

```python
plt.subplot(1,2,2)
sns.histplot(df['Height'], kde = True, color = 'green')
plt.title('Height Distribution (Continuous Data)')
```

```python
plt.tight_layout()
```

```python
plt.show()
```

# Univariate Analysis

```python
plt.figure(figsize = (10,4))
```

```python

plt.subplot(1,2,1)
sns.countplot(x='Gender',data = df, palette = 'Set1')
plt.title('Gender Count categorical data')
```

```python
plt.subplot(1,2,1)
sns.countplot(x='Smoker',data = df, palette = 'Set2')
plt.title('Smoker Count categorical data')
```

```python
plt.tight_layout()
plt.show()
```

# Bivariate Analysis

```python
sns.scatterplot(x='Age', y ='Height', hue = 'Gender', data = df)
plt.title('Scatter plot of age vs height' )
plt.show()
```

# Outlier detection - Boxplot

```python
plt.figure(figsize =  (8,6))
```

```python
sns.boxplot(x='Gender', y='Height',data = df)
plt.title('Boxplot of height by gender')
plt.show()
```

# Feature Engineering

```python
# Lets create new feature BMI

df['BMI'] = df['Weight']/(df['Height']/100)**2

print('Data with new BMI feature \n',df)
```

# Visualisation of the new feature

```python
sns.histplot(df['BMI'], kde = True, color = 'purple')
plt.title('BMI Distribution')
plt.show()
```
