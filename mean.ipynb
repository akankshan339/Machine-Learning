```python
#Detecting outlier in the dataset using different methods such as the Z-score, Interquartile Range (IQR), and Tukey’s Fences
#these methods identify data points that significantly differ from others in the dataset, improving data analysis and accuracy
```

```python
#Z-score method uses standard deviation to determine outliers
#Calculated z-score > threshold is considered an outlier
#Threshold generally lies between 2 to 3
```

```python
outliers =[]
threshhold = 3;
for x in data:
    z_score = (x - mean)/sd       #put actual mean
    if abs(z_score) > threshhold:
        outliers.append(x)
```

```python
print("outliers is ", outliers)
```

Method 2: Interquartile Range (IQR)

```python
#find IQR of data by subtracting q1 from q3 .If the value/data point is more than 1.5 times the iqr it will be considered an outlier.
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
```

```python
print(q1)
print(q3)
```

```python
iqr = q3 - q1
threshold = 1.5 * iqr
outliers = np.where((data < q1 - threshold) | (data > q3 + threshold))
#calculate percentile of 25th and 75th set threshold to 1.5 times iqr
```

```python
print(threshold)
```

```python
print(iqr)
```

```python
print(outliers)
```

```python
print("Outliers of array ",data,"is : \n", data[outliers])
```
