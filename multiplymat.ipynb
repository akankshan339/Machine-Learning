```python
import numpy as np
```

```python
arr1 = np.array([[1,2,3], [4,5, 6],[1,2,1]])
```

```python
arr2 = np.array([[1,1,1], [0,1,0],[1,5,3]])
```

```python
print(arr1)
```

```python
print(arr2)
```

```python
arr3 = np.multiply(arr1, arr2)
```

```python
print (arr3)
```

```python
#calculating dot product
arr3 = np.dot(arr1, arr2)
```

```python
print (arr3)
```

```python
#calculating cross product
arr3 = arr1@arr2
```

```python
print (arr3)
```

```python
sum = np.add(arr1, arr2)
```

```python
print (sum)
```

```python
sub = np.subtract(arr1, arr2)
```

```python
print (sub)
```

```python
div = np.divide(arr2, arr1)
```

```python
print (div)
```
